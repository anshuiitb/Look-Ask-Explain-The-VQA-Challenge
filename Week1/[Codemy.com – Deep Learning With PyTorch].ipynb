```python
import torch.nn as nn
import torch.nn.functional as F
```

```python
#create a Model Class that inherits nn.Module

class Model(nn.Module):
  def __init__(self , in_features=4 , h1=32 , h2=16 , h3=8 , out_features = 3):
    super().__init__()
    self.fc1 = nn.Linear(in_features , h1)
    self.fc2 = nn.Linear(h1 , h2)
    self.fc3 = nn.Linear(h2 , h3)
    self.out = nn.Linear(h3 , out_features)

  def forward(self,x):
    x=F.relu(self.fc1(x))
    x=F.relu(self.fc2(x))
    x=F.relu(self.fc3(x))
    x=self.out(x)
    return x

```

```python
torch.manual_seed(41)
model = Model()
```

```python
from sklearn.model_selection import train_test_split
X_train , X_test , y_train , y_test = train_test_split(X , y , test_size = 0.2 , random_state =41)
```

```python
X_train = torch.FloatTensor(X_train)
X_test = torch.FloatTensor(X_test)
y_train = torch.LongTensor(y_train)
y_test = torch.LongTensor(y_test)




# PyTorch layers like nn.Linear expect float32

# If X is int, youll get a dtype mismatch error
#  What CrossEntropyLoss expects:

# Inputs (logits)  FloatTensor

# Targets (labels)  LongTensor

# Labels must be:

# Class indices, not one-hot

# Example: 0, 1, 2 (not [0,1,0])
```

```python
criterion = nn.CrossEntropyLoss()

optimizer = torch.optim.Adam(model.parameters() , lr = 0.01)
```

```python
print(model.parameters)
```

```python
epochs = 100
losses = []
for i in range(epochs):
  y_pred = model.forward(X_train)
  loss = criterion(y_pred , y_train)

  losses.append(loss.detach().numpy())

  if i%10 == 0:
    print(f"epoch : {i} , loss : {loss}")

  optimizer.zero_grad()
  loss.backward()
  optimizer.step()
```

```python
plt.plot(range(epochs) , losses)
plt.ylabel("loss")
plt.xlabel("epoch")
```

```python
class CNNModel(nn.Module):
  def __init__(self):
    super().__init__()
    self.conv1 = nn.Conv2d(in_channels=1 , out_channels=6 , kernel_size=3 , stride=1)
    self.conv2 = nn.Conv2d(in_channels=6, out_channels=16 , kernel_size=3 , stride=1)

    self.fc1 = nn.Linear(5*5*16,120)
    self.fc2 = nn.Linear(120,84)
    self.fc3 = nn.Linear(84,10)

  def forward(self,x):
    x = F.relu(self.conv1(x))
    x = F.max_pool2d(x,2,2)
    x = F.relu(self.conv2(x))
    x = F.max_pool2d(x,2,2)
    x = x.view(-1,5*5*16)
    x = F.relu(self.fc1(x))
    x = F.relu(self.fc2(x))
    x = self.fc3(x)
    return F.log_softmax(x,dim=1)
```

```python
torch.manual_seed(41)
model = CNNModel()
```

```python
model
```

```python
criterion = nn.CrossEntropyLoss()
optimiser = torch.optim.Adam(model.parameters() , lr = 0.01)
```

```python
train_losses =[]
test_losses =[]
train_correct =[]
test_correct =[]
epochs = 5
import time
start_time = time.time()


for i in range(epochs):
  trn_corr=0
  tst_corr=0
  for b , (X_train , y_train) in enumerate(train_loader):
    b+=1
    y_pred = model(X_train)
    loss = criterion(y_pred , y_train)
    predicted = torch.max(y_pred.data , 1)[1]

    batch_corr = (predicted == y_train).sum()
    trn_corr += batch_corr

    optimiser.zero_grad()
    loss.backward()
    optimiser.step()





    if b%600 == 0:
      print(f"epoch : {i} , batch : {b} , loss : {loss.item()}")

  train_losses.append(loss)
  train_correct.append(trn_corr)


  with torch.no_grad():
    for b , (X_test , y_test) in enumerate(test_loader):
      y_val=model(X_test)
      predicted = torch.max(y_val.data , 1)[1]
      ts_corr = (predicted == y_test).sum()
      tst_corr += ts_corr



  loss = criterion(y_val , y_test)
  test_losses.append(loss)
  test_correct.append(tst_corr)



current_time = time.time()





total_time = current_time - start_time
print(f"total_time is :{total_time}")
```
